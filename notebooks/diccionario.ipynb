## Descripción Detallada

### Identificadores y Transacciones
    TransactionID: Identificador único de cada transacción. Este campo es clave para unir con el archivo train_identity.csv.isFraud: Indicador binario de si la transacción es fraudulenta (1) o no (0).

### Tiempo y Montos
    TransactionDT: Tiempo en segundos desde un punto de referencia (usado para crear características temporales).TransactionAmt: Monto de la transacción.ProductosProductCD: Código del producto involucrado en la transacción (A, B, C, D, W).

### Información de Tarjeta
    card1 - card6:
    card1: ID del emisor de la tarjeta.
    card2: Código de país del emisor de la tarjeta.
    card3: Clase de la tarjeta de crédito.
    card4: Marca de la tarjeta (por ejemplo, Visa, Mastercard).
    card5: Región del emisor de la tarjeta.
    card6: Tipo de tarjeta (débito, crédito).

### Dirección y Distancias
    addr1, addr2: Dirección de facturación.
    dist1, dist2: Distancia desde la dirección de facturación a la dirección de envío.

### Dominios de Correo Electrónico
    P_emaildomain: Dominio del correo electrónico del pagador.
    R_emaildomain: Dominio del correo electrónico del receptor.

### Características Derivadas de Transacciones
    C1 - C14: Características derivadas de las transacciones. Estas pueden incluir frecuencias de uso, cantidades totales y otros patrones basados en el historial de transacciones del usuario.

### Variables de Tiempo
    D1 - D15: Variables de tiempo. Por ejemplo, 
    D1 puede representar el número de días desde la última transacción. Estas características ayudan a capturar patrones temporales en las transacciones.

### Indicadores de Emparejamiento
    M1 - M9: Indicadores binarios que muestran si ciertos detalles de la transacción coinciden o no. Por ejemplo, si la dirección de envío y facturación coinciden.

### Variables Derivadas
    V1 - V339:Variables derivadas de la identidad y comportamiento. Estas son características complejas creadas a partir de datos de identidad, detalles del dispositivo y patrones de comportamiento
